```python
import random


def print_hangman(turns):
    stages = [
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     / \\
        """,
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |     / 
        """,
        """
           --------
           |      |
           |      O
           |     \\|/
           |      |
           |      
        """,
        """
           --------
           |      |
           |      O
           |     \\|
           |      |
           |     
        """,
        """
           --------
           |      |
           |      O
           |      |
           |      |
           |     
        """,
        """
           --------
           |      |
           |      O
           |    
           |      
           |     
        """,
        """
           --------
           |      |
           |      
           |    
           |      
           |     
        """
    ]
    print(stages[turns])


def play_again():
    while True:
        choice = input("Do you want to play again? (y/n): ").lower()
        if choice == 'y':
            return True
        elif choice == 'n':
            return False
        else:
            print("Invalid input. Please enter 'y' or 'n'.")


def choose_topic():
    topics = ['movies', 'sports', 'programming languages', 'novels', 'history', 'comic books', 'video games']
    while True:
        print("Choose a topic:")
        for i, topic in enumerate(topics, 1):
            print(f"{i}. {topic}")
        choice = input("Enter the number corresponding to the topic: ")
        if choice.isdigit() and 1 <= int(choice) <= len(topics):
            return topics[int(choice) - 1]
        else:
            print("Invalid input. Please enter a number from the given options.")


def main():
    print("Welcome to the Word Guessing Game!")

    while True:
        choice = input("Do you want to play the game? (y/n): ").lower()
        if choice == 'n':
            break
        elif choice == 'y':
            topic = choose_topic()
            print("You have chosen the topic:", topic)

            words = {
                'movies': ['avengers', 'inception', 'jaws', 'titanic', 'matrix', 'avatar'],
                'sports': ['soccer', 'tennis', 'basketball', 'cricket', 'swimming', 'volleyball'],
                'programming languages': ['python', 'java', 'csharp', 'javascript', 'ruby', 'go'],
                'novels': ['pride', 'prejudice', 'harry', 'potter', 'gatsby', 'mockingbird'],
                'history': ['rome', 'egypt', 'greece', 'china', 'india', 'mayans'],
                'comic books': ['superman', 'spiderman', 'batman', 'wonderwoman', 'hulk', 'thor'],
                'video games': ['mario', 'minecraft', 'fortnite', 'zelda', 'pokemon', 'fallout']
            }

            word = random.choice(words[topic]).lower()
            word_length = len(word)
            guesses = set()
            turns = 6

            print("Guess the characters:")
            print("_ " * word_length)

            while turns > 0:
                print_hangman(turns)

                guessed_word = ""
                for char in word:
                    if char in guesses:
                        guessed_word += char + " "
                    else:
                        guessed_word += "_ "

                print(guessed_word)

                if "_" not in guessed_word:
                    print("Congratulations! You won!")
                    break

                guess = input("Guess a character: ").lower()

                if len(guess) != 1 or not guess.isalpha():
                    print("Invalid input. Please enter a single letter.")
                    continue

                if guess in guesses:
                    print("You already guessed that letter. Try again.")
                    continue

                guesses.add(guess)

                if guess not in word:
                    turns -= 1
                    print("Wrong guess!")
                    print("You have", turns, "more guesses.")

                    if turns == 0:
                        print("Sorry, you lost. The word was", word)
                        break

            print("\n")  # Gap after losing the game

        else:
            print("Invalid input. Please enter 'y' or 'n'.")

        if not play_again():
            break


if __name__ == "__main__":
    main()
```
